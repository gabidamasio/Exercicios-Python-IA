```python
import random
from collections import defaultdict

def train(text, n):
    words = text.split()
    data = defaultdict(list)

    for i in range(len(words)-n):
        context = tuple(words[i:i+n])
        next_word = words[i+n]
        data[context].append(next_word)

    lwords = [word for context, words in data.items() for word in words]
    return data, lwords

def predict(data, context, lwords):
    if context not in data or not data[context]:
        return random.choice(lwords)
    return random.choice(data[context])

def main():
    text = "A melhor parte do meu dia é o seu sorriso. Estar com você foi a decisão mais fácil que tomei na vida e me sinto a pessoa mais sortuda do mundo todos os dias."
    n = 2

    data, lwords = train(text, n)

    while True:
        input_text = input("Digite uma frase para prever a próxima palavra (ou digite 'sair' para encerrar): ")
        if input_text.lower() == 'sair':
            break
        context_words = input_text.split()[-n:]
        context = tuple(context_words)
        predicted_word = predict(data, context, lwords)
        print("Próxima palavra prevista:", predicted_word)

if __name__ == "__main__":
    main()
```
