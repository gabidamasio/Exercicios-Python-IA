```python
!pip install nltk
!pip install beautifulsoup4
nltk.download('averaged_perceptron_tagger')

import nltk
from nltk.tokenize import WhitespaceTokenizer
import requests
from bs4 import BeautifulSoup
from collections import Counter
from nltk.tokenize import word_tokenize


def load_text(file_path):
    with open(file_path, 'r', encoding='utf-8') as file:
        text = file.read()
    return text

def save_webpage_text(url, file_path):
    response = requests.get(url)
    if response.status_code == 200:
        html_content = response.text
        soup = BeautifulSoup(html_content, 'html.parser')
        text = soup.get_text()
        with open(file_path, 'w', encoding='utf-8') as file:
            file.write(text)
        print("O texto da página foi salvo com sucesso em", file_path)
    else:
        print("Falha ao obter conteúdo da página:", response.status_code)

url = "https://www.avvi.com.br/"
file_path = "corpus.txt"
save_webpage_text(url, file_path)

text = load_text(file_path)
print("Conteúdo do arquivo:")
print(text)

class TokenizerI:
    def tokenize(self, text):
        raise NotImplementedError("Método 'tokenize' não implementado")

class WhitespaceTokenizerCustom(TokenizerI):
    def tokenize(self, text):
        tokenizer = WhitespaceTokenizer()
        tokens = tokenizer.tokenize(text)
        return tokens

tokenizer = WhitespaceTokenizerCustom()
tokenized_text = tokenizer.tokenize(text)
print("\nTexto tokenizado:")
print(tokenized_text)

num_tokens = len(tokenized_text)
print("\nNúmero de tokens:", num_tokens)

selected_words = ["Shorts", "Leggings e Calças", "Tops", "Conjuntos", "Macacão", "Café"]
tokens = word_tokenize(text.lower())
tagged_words = nltk.pos_tag(tokens)
word_tag_freq = {word: Counter() for word in selected_words}

for word, tag in tagged_words:
    for selected_word in selected_words:
        if selected_word.lower() in word.lower():
            word_tag_freq[selected_word][tag] += 1
print("Frequências das tags para as palavras selecionadas:")

for word in selected_words:
    print(f"Palavra: {word}")
    print("   Frequências das tags:")
    for tag, freq in word_tag_freq[word].items():
        print(f"      Tag: {tag}, Frequência: {freq}")
```
